# boxplot_example

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame()
for i, (mi, std) in enumerate(zip(np.arange(1, 6), np.sqrt(np.arange(5, 0, step=-1)))):
    df['x'+str(1+i)]=np.random.normal(mi, std, size=4000)
df['group']=np.array(list(np.arange(1, 1+4))*(1+df.shape[0]//4))[:df.shape[0]]
df
```

```python
plt.style.use('bmh')
fig, axes=plt.subplots(2, 2, figsize=(16, 9))
axes=axes.reshape(4)

for i, k in enumerate(np.arange(1, 1+4)):
    df_k=df[df.columns[:-1].to_list()].loc[df.group==k]
    axes[i].boxplot([df_k[column].values for column in df_k.columns.to_list()],
                    showmeans=True, vert=True, labels=df_k.columns.to_list())
    axes[i].set_title('Boxplots of group {:,.0f}'.format(k))

plt.show()
plt.close()
```
